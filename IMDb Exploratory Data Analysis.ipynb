```python
## Pandas
import pandas as pd
## Numpy
import numpy as np

import json, os, time

import tmdbsimple as tmdb

from tqdm.notebook import tqdm_notebook

import gzip
```

```python
# Open
tmdb_20 = pd.read_csv("data/final_tmdb_data_2000.csv.gz",low_memory = False)
tmdb_20
```

```python
# Open
tmdb_21 = pd.read_csv("data/final_tmdb_data_2001.csv.gz",low_memory = False)
tmdb_21
```

```python
frames = [tmdb_20,tmdb_21]
imdb = pd.concat(frames)

imdb.head()
```

```python
imdb.info()
```

```python
imdb_fil = imdb.loc[(imdb.revenue > 0) | (imdb.budget > 0)]

imdb_fil.info()
```

```python
imdb_fil.to_csv("data/tmdb_results_combined.csv.gz", compression='gzip', index=False)
```

### number of movies per certification

```python
imdb_fil['certification'].value_counts()
```

### average revenue per certification category

```python
revenue_per_cert = imdb_fil.groupby('certification')['revenue'].mean()

revenue_per_cert
```

### average budget per certification category

```python
budget_per_cert = imdb_fil.groupby('certification')['budget'].mean()

budget_per_cert
```
