```python
## Pandas
import pandas as pd
## Numpy
import numpy as np

import json, os, time

import tmdbsimple as tmdb

from tqdm.notebook import tqdm_notebook

import gzip
```

```python
# Open
tmdb_20 = pd.read_csv("data/final_tmdb_data_2000.csv.gz",low_memory = False)
tmdb_20
```

```python
# Open
tmdb_21 = pd.read_csv("data/final_tmdb_data_2001.csv.gz",low_memory = False)
tmdb_21
```

```python
frames = [tmdb_20,tmdb_21]
imdb = pd.concat(frames)

imdb.head()
```

```python
imdb.info()
```

```python
rev_fil = imdb['revenue'] > 0

results_rev = imdb[rev_fil]
print(f'Total valid revenue movie count is {results_rev.shape[0]}')
```

```python
bud_fil = imdb['budget'] > 0

results_bud = imdb[bud_fil]
print(f'Total valid revenue movie count is {results_bud.shape[0]}')
```

```python
imdb['certification'].value_counts()
```

```python
revenue_per_cert = imdb.groupby('certification')['revenue'].mean()

revenue_per_cert
```

```python
budget_per_cert = imdb.groupby('certification')['budget'].mean()

budget_per_cert
```
